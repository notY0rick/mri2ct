```python
import nibabel as nib
import numpy as np
```

```python
folder_path = "/content/drive/MyDrive/cs280_sp2023/project/1BC074/"
ct = folder_path + "ct.nii.gz"
mri = folder_path + "mr.nii.gz"
mask = folder_path + "mask.nii.gz"
```

```python
def nii2np(path):
    img = nib.load(path)
    return np.asanyarray(img.dataobj)
```

```python
mri_np = nii2np(mri)
print(mri_np.shape)
```

```python
ct_np = nii2np(ct)
print(ct_np.shape)
```

```python
mask_np = nii2np(mask)
print(mask_np.shape)
```

```python
from array2gif import write_gif
import cv2

def create_gif(img_np, title="test"):
    frames = []
    for i in range(img_np.shape[2]):
        frame = img_np[:, :, i][None]
        frame = cv2.normalize(frame, None, alpha = 0, beta = 255, norm_type = cv2.NORM_MINMAX, dtype = cv2.CV_32F)
        frame = frame.astype(np.uint8)
        frame = np.concatenate([frame, frame, frame], axis=0)
        frame = np.clip(frame, 0, 255)
        frames.append(frame)
    write_gif(frames, "/content/drive/MyDrive/cs280_sp2023/project/" + title + '.gif', fps=5)

create_gif(mri_np, title='mri')
```

```python
create_gif(ct_np, title='ct')
```

```python
create_gif(mask_np, title='mask')
```
